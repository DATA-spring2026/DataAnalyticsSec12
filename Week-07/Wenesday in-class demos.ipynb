### Adding a column to a DataFrame

```python
inventory_df2
```

```python
inventory_df2.insert(2, "price", [1.99, 3.49, 13.50])

inventory_df2
```

```python
inventory_df2.insert(1, "price2", [1.99, 3.49, 13.50])

inventory_df2
```

```python
# We can also drop rows / columns
# to drop rows, use axis = 0
# to drop columns, use axis = 1

inventory_df2.drop("price2", axis=1, inplace=True)  # default is inplace=False

inventory_df2
```

```python
inventory_df2.loc["bell pepper"]    # using .loc[]
```

```python
inventory_df2.loc[["bell pepper", "ground beef"]]    # using .loc[[]]  with additional set of brackets
```

```python
inventory_df2.iloc[[1, 2]]    # using .iloc[[]] to retrieve by index number position
```

```python
inventory_df2.loc["organic eggs"] = [1234, 105, 10.99]

inventory_df2
```

## A note on dropping and deduplicating row data

.drop() method can be used to drop rows (axis = 0)

.dropna() will drop undefined values

.drop_duplicates() will remove duplicate record

For each of these methods, if you want to save that change to the original underlying DataFrame, use inplace=True (default is to return a copy, inplace=False)

## Using groupby

```python
sales_det = pd.DataFrame({ "prod_num": [658, 2476, 658, 2476, 3911, 3911], "price": [7.89, 5.49, 7.89, 5.49, 6.25, 6.25]})

sales_det
```

```python
sales = sales_det.groupby(['prod_num']).sum()   # total revenue

sales
```
