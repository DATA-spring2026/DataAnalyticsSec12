## Using where()

```python
arr
```

```python
print(np.where(arr >= 3))  # where requires a condition (using <, >, ==, etc.)
```

```python
print(np.where(arr % 3 == 0, arr, np.nan)) 
# first argument is the condition
# second argument is what to return if condition IS met
# third argument is what to return if condition IS NOT met

# this pattern of condition is often used to identify even numbers (variable_name % 2 == 0)
```

## Working with NumPy's Random

```python
# We saw randint() generates random integers

print(np.random.randint(10))

print(np.random.randint(10, 100))

print(np.random.randint(10, 100, size = (2, 5)))
```

```python
# rand() is going to generate a random float btwn 0 and 1

print(np.random.rand())

print(np.random.rand(2, 3))
```

```python
# NumPy choice() can be used to sample from an array

print(np.random.choice(product_num))

print(np.random.choice(product_num, size = (2, 3)))
```

```python
pets = ['cat','dog','bunny','snake','fish']

sales_this_week = np.random.choice(pets, p = [.2, .1, .6, 0, .1], size = 10)

print(sales_this_week)
```

## Manipulating arrays

```python
# shuffle and permutation work similarly
# shuffle is going to reorder the original array (in-place)
# permutation is going to return a new array with values in a shuffled order

grocery_prices = np.array([2.30, 4.99, 3.75, 11.00, 1.99])

print(grocery_prices)

print(np.random.permutation(grocery_prices))

print(grocery_prices)
```

```python
# using shuffle

print(grocery_prices)

print(np.random.shuffle(grocery_prices))  # shuffles the values in place, doesn't return a value

print(grocery_prices)
```

## NumPy ufuncs

```python
# example using plain old multiplication

discount = np.array([.1, .15, .1, .15, .2])

print(grocery_prices * discount)
```

```python
print(np.multiply(grocery_prices, discount))  # a lot faster in computer processing
```

```python
print(np.add(grocery_prices, grocery_prices))
```

```python
print(np.add(grocery_prices, 2))
```

```python
new_prices = np.array([1.2, 1.5, 1.9, 2.2, 2.5, 2.9, 3.5, 4.5])

# if halfway between values (e.g. 1.5, 2.5) then NumPy rounds to the nearest EVEN integer

print(np.round(new_prices))
```

```python
# sum() lets you add up the elements in an array

print(np.sum(grocery_prices))
```

```python
groc_price_2 = np.array([[2, 3, 4, 5, 10], [6, 7, 2, 1, 4]])

print(groc_price_2)
```

```python
print(np.sum(groc_price_2))
```

```python
print(np.sum(groc_price_2, axis = 1))  # sum can work on a specified axis
```

```python
# cumulative sum provides a rolling total / rolling sum of values

print(np.cumsum(groc_price_2))
```

```python
# similarly there's prod() to multiply all values in array

print(np.prod(groc_price_2))
```

```python
print(np.cumprod(groc_price_2))
```

```python
# diff() successive different elements

print(groc_price_2)
print(np.diff(groc_price_2))
```

```python
print(np.diff(groc_price_2, n=2))
```

```python
print(np.diff(groc_price_2, n=3))
```

```python
print(np.diff(groc_price_2, n=4))
```

```python
# another rounding example

for_rounding = np.array([1.5, 2.5, 3.5, 4.5, 5.5])

print(for_rounding)
print(np.round(for_rounding))
```
