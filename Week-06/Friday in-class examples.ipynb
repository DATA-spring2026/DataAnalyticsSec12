#### Adding a lambda:

```python
input_file = "sample.txt"
output_file = "sales_data2.csv"

revenue_per_scoop = lambda scoops, revenue: revenue / scoops   # this lambda is creating a little calculator as revenue_per_scoop
```

```python
with open(input_file) as f:
    lines = f.readlines()
```

```python
with open(output_file, "w") as o:
    for i in lines:
        if "," in i:
            parts= i.strip().split(",")
            if parts[0] == 'Flavor':
                o.write(i.strip() + ",Revenue Per Scoop\n")
            else: 
                scoops = int(parts[1])
                revenue = float(parts[3].replace("$", ""))
                rps = revenue_per_scoop(scoops, revenue)      # this is where we're using the lambda function
                o.write(f"{i.strip()},${rps:.2f}\n")

print("Output complete")
```

# Error handling

```python
try:
    first_n = float(input('% to calculate: '))
    answer = first_n / 100
except ValueError:
    print('Invalid input! Must be a number')
else:
    print(answer)
finally:
    print('Run again to try another number')
```

```python
# if we didn't use try:

first_n = float(input('% to calculate (as decimal value): '))
answer = first_n / 100

print(answer)
```

### Using try/except with file handling

```python
input_file = "robot_army.txt"
output_file = "sales_data3.csv"

try:
    with open(input_file) as f:
        lines = f.readlines()

except FileNotFoundError:
    print(f"Error: '{input_file}' not found.")

else:
    with open(output_file, "w") as out:
        for line in lines:
            if "," in line:
                out.write(line)
        print(f"Sales table written to '{output_file}'.")
```
