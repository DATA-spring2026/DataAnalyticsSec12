```python
# Create a variable for your starting bank balance, another that sets your savings goal, and a third with your weekly savings amount.

starting_balance = int(input('What is your bank balance? '))    # This was where my error was: I changed the code but didn't actually re-run the cell

savings_goal = int(input('What is your savings goal? '))

weekly_savings = int(input('What are you saving each week? '))
```

```python
# Use a while loop to compare your bank balance to your savings goal, if you haven’t met your goal yet, add the weekly savings amount to your bank 
# balance. For each loop, print the statement, “This week my balance increased to ___.” Once your savings goal is met, print the statement, “Goal 
# met! My current balance is ___.”

# print(starting_balance)  # validating the starting balance

# print(savings_goal > starting_balance)  # validating the condition

while savings_goal > starting_balance:
    starting_balance += (weekly_savings)
    print(f'This week\'s savings: {weekly_savings}')
    print(f'This week my balance increased to {starting_balance}')
```
