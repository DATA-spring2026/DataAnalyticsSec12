### Creating a basic while loop

Use a while loop to compare your bank balance to your savings goal, if you haven’t met your goal yet, add the weekly savings amount to your bank balance. For each loop, print the statement, “This week my balance increased to ___.” Once your savings goal is met, print the statement, “Goal met! My current balance is ___.”

```python
# print(starting_balance)  # validating the starting balance

# print(savings_goal > starting_balance)  # validating the condition

while savings_goal > starting_balance:
    starting_balance += (weekly_savings)
    print(f'This week\'s savings: {weekly_savings}')
    print(f'This week my balance increased to {starting_balance}')
    print()
```

### Adding conditional logic

```python
# resetting the starting values for each variable

starting_balance = 850
savings_goal = 1100
weekly_savings = 100

# alternatively, using input:
# starting_balance = int(input('What is your bank balance? '))
# savings_goal = int(input('What is your savings goal? '))
# weekly_savings = int(input('What are you saving each week? '))
```

Try adding additional logic to your loop:

**a)** If your balance is more than halfway to your goal, print the message, “Almost there! This week my balance is up to ___.”

```python
while savings_goal > starting_balance:
    starting_balance += weekly_savings
    print(starting_balance)
    if starting_balance > (savings_goal / 2):
        print(f'Almost there! This week my balance is up to {starting_balance}')
    else:
        print(f'This week my balance increased to {starting_balance}')
```

**b)** If your balance is at least 75% of your goal, add a calculation to buy yourself a little treat. Print the message, “So close! After treating myself, my balance is up to ___.”

```python
# resetting the starting values AGAIN for each variable
# this time using a lower starting point for starting_value

starting_balance = 150
savings_goal = 1100
weekly_savings = 100

# alternatively, using input:
# starting_balance = int(input('What is your bank balance? '))
# savings_goal = int(input('What is your savings goal? '))
# weekly_savings = int(input('What are you saving each week? '))
```

```python
while savings_goal > starting_balance:
    starting_balance += weekly_savings
    print(starting_balance)
    if starting_balance > (savings_goal * .75):
        starting_balance -= 5  # nice-to-have
        print(f'So close! After treating myself, my balance is up to {starting_balance}')
    elif starting_balance > (savings_goal / 2):
        print(f'Almost there! This week my balance is up to {starting_balance}')
    else:
        print(f'This week my balance increased to {starting_balance}')
```
