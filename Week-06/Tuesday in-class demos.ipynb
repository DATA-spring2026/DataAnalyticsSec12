### Using the format method to fill in a placeholder

```python
txt1 = "The ribeye is {price:.2f} dollars, yikes"
print(txt1.format(price = 49))
```

```python
receipt_msg = "Your total today is {total:.2f} - thank you!"

food_bill = {"entree1": 18.00, "entree2": 23.00, "drink1": 4.00, "drink2": 8.00}

subtotal = sum(food_bill.values())

tax = subtotal * .08

print(receipt_msg.format(total = subtotal + tax))
```

### Working with the range function

```python
# one argument

for num in range(11):
    print(num)
```

```python
# two arguments

for num in range(1, 11):
    print(num)
```

```python
for num in range(50, 61):
    print(num)
```

```python
# using three arguments

for i in range(50, 61, 3):
    print(i)
```

```python
# using values produced by range() to create some groupings

survey_bands = []
counter = 1

for i in range (50, 61, 3):
    survey_bands.append(f'Group {counter}: {i} through {i+2}')
    counter += 1

print(survey_bands)
```

```python
# change the definition of burrito to return the string

def burrito():
    return "let's make a burrito"
```

```python
burrito()
```

```python
print(burrito())
```

```python
# redefining burrito() to include some stuff in our burrito

def burrito(protein, rice, beans, salsa, cheese):
    bur_order = []
    # protein = 'chicken'
    bur_order.append(protein)
    bur_order.append(rice)
    bur_order.append(beans)
    bur_order.append(salsa)
    bur_order.append(cheese)
    return bur_order
```

```python
# hard-coding the arguments that we want to supply

burrito('beef', 'no rice', 'black beans', 'pico de gallo', 'no cheese')
```

```python
burrito(
    input("What kind of protein do you want? "),
    input("Rice?"),
    input("Beans?"),
    input("Any salsa?"),
    input("Cheese?")
)
```

```python
# burrito() requires arguments for the parameters: protein, rice, beans, salsa, cheese

burrito(rice='brown rice',cheese='yes cheese', protein='grilled veg', beans='black beans', salsa='pico de gallo')
```
