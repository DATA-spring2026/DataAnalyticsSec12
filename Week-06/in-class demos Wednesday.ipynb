```python
def greeting_kids(*args):
    return args
```

```python
# keyword arguments

def rels(rel, name, **kwargs):
    kwargs['relationship'] = rel
    kwargs['name'] = name
    return kwargs
```

```python
person = rels('cousin', 'Bea', car='Toyota Corolla', nickname='gooseface')

print(person)
```

```python
# another example

def food_card_order(item, **kwargs):
    kwargs['order item'] = item
    return kwargs
```

```python
order1 = food_card_order('hot dog', drink='iced tea', side='fries')

print(order1)
```

## Lambda functions

```python
x = lambda a, b: a * b
```

```python
print(x(5, 6)) 
```

```python
for i in range(5):
    print(x(i, 10))
```

```python
for i in range(5):
    print((lambda y, z: y * z)(i, 10))
```

```python
def discount_calc1(price, pct_off):
    return price * (1 - pct_off)

print(discount_calc1(10, .3))
```

```python
def discount_calc2(pct_off):
 return lambda price: price * (1 - pct_off)
```

```python
pct_off20 = discount_calc2(.2)  # price not yet defined

pct_off30 = discount_calc2(.3)  # price not yet defined

base_price = float(input("Supply base price for item: "))

print(f'20% off: {pct_off20(base_price)}, 30% off:{pct_off30(base_price)}')
```

```python
# using lambda function with the map function

prices = [10, 25, 8, 40, 15]

discounted = list(map(lambda p: p * 0.8, prices))  # map() applies a function across a set of items
discounted2 = map(lambda p: p * 0.8, prices)  # without wrapping in list()

print(discounted) # [8.0, 20.0, 6.4, 32.0, 12.0]
print(discounted2)
```

```python
expensive = list(filter(lambda p: p > 12, prices))  # filter applies a condition across a collection and returns the items that match True
print(expensive)   # [25, 40, 15]
```

## Working with generator functions

```python
def greeting():
    yield "Hello"
    yield "How are you?"
    how_doing = (input("Your reply: ").lower())
    if how_doing == "good" or how_doing == "great":
        yield "That’s good. What’s new?"
    else:
        yield "Hey, how about this weather?"
```

```python
for x in greeting():
    print(x)
```

```python
for x in greeting():
    print(x)
```

```python
# using locally-scoped variables in a generator function

x = 'peanut butter'

def four_count():
    x = 0
    while x < 4:
        yield x
        x += 1
```

```python
for i in four_count():
    print(i)
```

## Working with Classes

```python
class SimpleDog:
    '''Simple model for a dog'''     # docstring description
    
    def __init__(self, name, age, he_she, fuzzy='fuzzy'):
        self.name = name
        self.age = age
        self.he_she = he_she
        self.fuzzy = fuzzy
        
    def sit(self): # this is a method
        return f'{self.name} is now sitting'
```

```python
my_dog = SimpleDog('Bobo', 12, 'He')  # creates an instance of SimpleDog
```

```python
print(f"My dog's name is {my_dog.name}")
```

```python
print(f'{my_dog.he_she} is {my_dog.age} years old')
```

```python
my_dog.sit()
```

```python
coreys_dog = SimpleDog('Doomsday', 1, 'He')
```

```python
print(f'Hey, guys, doesn\'t {coreys_dog.name} look awfully like a cat?')
```

```python
coreys_dog.sit()
```
