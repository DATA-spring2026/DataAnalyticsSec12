## Quick example of a loop

```python
for i in fruits:
    print (i)
```

```python
for n, i in enumerate(fruits):
    print(n, i)
```

```python
for n, i in enumerate(fruits, start=1):
    print(f'{n}. {i}')
```

```python
for peanutbutter, jelly in enumerate(fruits, start=1):
    print(f'{peanutbutter}. {jelly}')
```

```python
# using find

fan_statement = "my favorite muppets are rizzo, gonzo, miss piggy, and kermit"
if "kermit" in fan_statement:
    print(fan_statement.find("kermit"))
    print(fan_statement[fan_statement.find("kermit")-10:fan_statement.find("kermit")])
    print(fan_statement[fan_statement.find("kermit")-10:])
```

```python
# using rfind instead of find

fan_statement = "my favorite muppets are rizzo, gonzo, miss piggy, and kermit"
if "kermit" in fan_statement:
    print(fan_statement.rfind("kermit"))
    print(fan_statement[fan_statement.find("kermit")-10:fan_statement.rfind("kermit")])
    print(fan_statement[fan_statement.rfind("kermit")-10:])
```
