# Continuing with Classes

## Working with class methods

```python
class SimpleDog:
    '''
    Simple model for a dog
    Here is more descript of the dog. Also,
    I have other comments about the project...
    '''
    
    def __init__(self, name, age, he_she, fuzzy='fuzzy'):
        self.name = name
        self.age = age
        self.he_she = he_she
        self.fuzzy = fuzzy
        self.behavior = 'good'
        
    def sit(self): # this is a method
        return f'{self.name} is now sitting'

    def birthday(self):
        self.age += 1
        return f'Happy birthday, {self.name}! {self.he_she} is now {self.age}'

    def behaved(self):
        self.behavior = input(f"How is {self.name} behaving right now? ")
        print(f'{self.behavior.title()} dog!')
        self.behavior = 'good'
        return f'{self.he_she} is a {self.behavior} dog'
```

```python
luis_dog = SimpleDog('Lola', 8, 'She', 'short-haired')
```

```python
print(f'{luis_dog.name} is {luis_dog.fuzzy}')
print(luis_dog.birthday())
```

```python
luis_dog.behaved()
```

### Class inheritance -- creating a child class

```python
class Labrador(SimpleDog):
    ''' Dog breed is Labrador'''

    def __init__(self, name, age, he_she, fuzzy='fuzzy', swims='good swimmer'):
        super().__init__(name, age, he_she, fuzzy)
        self.swims = swims
```

```python
lab1 = Labrador('Pete', 6, 'He')
```

```python
print(lab1.birthday())
print(lab1.behaved())
print(lab1.sit())
```

```python
print(lab1.swims)
```

```python
# opening the file using a with block  -- best practice so you don't forget to close your files!

with open("sample.txt") as f:
    print(f.read(100))
```

```python
with open("sample.txt") as f:
    print(f.readline())
    print(f.readline())
```

```python
with open("sample.txt") as f:
    for i in range(5):
        print(f.readline())
```

```python
with open("sample.txt") as f:
    print(f.readlines())     # parses a file into a set of strings
```

```python
with open("sample.txt") as f:
    print(f.readlines(10))    # max number of characters retrieved, rounded up to the nearest whole line
```

```python
with open("sample.txt") as f:
    print(f.readlines(100))
    print(f.readlines(100))
```

```python
with open("sample.txt", "a") as f:
    f.write("Weather was sunny and mild")
```

```python
with open("sample.txt") as f:
    print(f.read())
```
